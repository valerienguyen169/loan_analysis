```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report


# Load the dataset
df = pd.read_csv("loan_approval_dataset.csv")
```

```python
# Summary statistics
print("\nSummary statistics:")
print(df.describe().T)
```

```python
# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Data visualization
# Histogram of loan_amount
plt.figure(figsize=(10, 6))
sns.histplot(df[' loan_amount'], bins=20, color='skyblue', kde=True)  # Remove the space before 'loan_amount'
plt.title('Distribution of Loan Amount')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Box plot of loan_amount by loan_status
plt.figure(figsize=(10, 6))
sns.boxplot(x=' loan_status', y=' loan_amount', data=df)
plt.title('Loan Amount by Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Loan Amount')
plt.grid(True)
plt.show()
```

```python
# Count plot of education level
plt.figure(figsize=(8, 6))
sns.countplot(x=' education', data=df, palette='Set2')
plt.title('Count of Applicants by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Exclude non-numeric columns and handle categorical variables
numeric_cols = df.select_dtypes(include=['number']).columns
df_numeric = df[numeric_cols]

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Count plot of loan_status by education level
plt.figure(figsize=(8, 6))
sns.countplot(x=' education', hue=' loan_status', data=df, palette='Set2')
plt.title('Loan Approval Status by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title='Loan Status', loc='upper right')
plt.show()
```
